```python

# Import nltk module  
import nltk  
# Download ntlk data  
# URL of the TOI's news article  
import pandas as pd
from newspaper.configuration import Configuration

# Set display options for wider visibility
pd.set_option('display.max_colwidth', None)  # Prevent column text truncation
pd.set_option('display.max_columns', None)  # Show all columns
pd.set_option('display.max_rows', None)     # Show all rows

sd=pd.read_json("newsapi_articles.json")
articles_title=[]
articles_text=[]
articles_date=[]
urls=[]
errors=[]

df=pd.DataFrame(sd)
config=Configuration()
config.browser_user_agent="Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"
i=0
list_of_urls=df["url"].tolist()
for url in list_of_urls:
    # Step 2: Fetch the webpage
    url=url.strip(" ")
    urlOfArticle = url 
    try:
    # Creating instance for the article  
        instanceOfArticle = Article(urlOfArticle, language="en",config=config) # en is for English  
    # Downloading the article piece from the program  
        instanceOfArticle.download()  
    # Parsing the article  

        instanceOfArticle.parse()  
        # Performing NLP on the article piece  
        # Extracting title of the article  
        articles_title.append( instanceOfArticle.title)  
        #print("n")  
        # Extracting texts from the article  
        articles_text.append( instanceOfArticle.text)
        if instanceOfArticle.publish_date:
            articles_date.append(instanceOfArticle.publish_date)
        else:
            articles_date.append("NaN")
        urls.append(url)
        i=i+1
        print(i)
        #print("n")  
        # Extract summary of the article  
        #print("Summary of the article piece:", instanceOfArticle.summary)  
        #print("n")  
        # Extracting important keywords of the article  
        #print("Important Keywords of the article:", instanceOfArticle.keywords)  
        #print("n")  
        #print ("We have successfully performed scrapping from the piece of article's link given in the code!")
    except Exception as e:
        errors.append((url,str(e)))
```

```python
import pandas as pd
import requests
import json
# Ollama API endpoint
OLLAMA_URL = "http://127.0.0.1:11434/api/chat"  # Default Ollama server URL

# Input/Output CSV file names
input_csv = "search_engine_intermediate_data_stopwords_removed.csv"   # Input CSV file with articles
output_csv = "search_engine_analyzed_data1.csv" # Output CSV file with extracted details

# Model name (your custom model)
MODEL_NAME = "llama3.2"


def analyze_article(article):
    """
    Send the article to Ollama for analysis and process streaming responses.
    """
    payload = {
        "model": MODEL_NAME,
        "messages": [
            {
                "role": "user",
                "content": f"Analyze the following article and extract: country, region, product name, transportation status, sentiment, and risk analysis (from -1 to 1).\n\n{article}"
            }
        ]
    }

    try:
        print("Analyzing article...")
        with requests.post(OLLAMA_URL, json=payload, stream=True) as response:
            response.raise_for_status()

            # Combine the content from streaming chunks
            full_response = ""
            for line in response.iter_lines():
                if line:  # Ignore keep-alive chunks
                    try:
                        json_chunk = json.loads(line)
                        if "message" in json_chunk and "content" in json_chunk["message"]:
                            full_response += json_chunk["message"]["content"]
                    except json.JSONDecodeError as e:
                        print(f"Error decoding chunk: {line} - {e}")

            print(f"Final response: {full_response}")
            return full_response

    except requests.exceptions.RequestException as req_err:
        print(f"Request error: {req_err}")
        return "Request error"
    except Exception as e:
        print(f"Error querying Ollama: {e}")
        return "Error"




# Step 1: Load the input CSV file
try:
    data = pd.read_csv(input_csv)
    print("Input file loaded successfully.")
except FileNotFoundError:
    print(f"Error: File {input_csv} not found.")
    exit()

# Step 2: Check for 'article' column
#if "article" not in data.columns:
 #   print("Error: Input CSV must have an 'article' column with the text to analyze.")
  #  exit()

# Step 3: Analyze each article




# Step 4: Add results to a new column and save to output CSV
#data['analysis']=analyze_article(data["summary"][3])
data['analysis'] = data['content'].apply(lambda x: analyze_article(x))

data.to_csv(output_csv, index=False)
print(f"Analysis results saved to {output_csv}")
```
